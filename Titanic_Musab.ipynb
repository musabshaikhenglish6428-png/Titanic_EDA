```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"/content/train.csv")
print(df)
```

```python
df.info()        # see columns, data types, missing values
df.describe()    # see stats for numerical columns
df.isnull().sum() # check missing values
```

```python
df['Age'].fillna(df['Age'].median())
df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'].fillna((df['Cabin'].mode()))
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

Overall Survival Count

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.histplot(x='Age', bins='10', kde= True)
plt.title('Age Distributon')
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # includes the passenger themselves
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()
```

```python
sns.kdeplot(df[df['Survived']==0]['Age'], label='Not Survived', shade=True)
sns.kdeplot(df[df['Survived']==1]['Age'], label='Survived', shade=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

Overall Survival:

Around 38% of passengers survived, 62% did not.

Survival by Gender:

Women had a much higher survival rate than men.

Survival by Class:

Passengers in 1st class survived more than those in 2nd and 3rd class.

Family Size:

Small families (2â€“4 members) had slightly higher survival chances.

Passengers alone or in very large families survived less.

Fare:

Higher fare passengers (usually 1st class) had higher survival rates.

Lower fare passengers (usually 3rd class) survived less.

Age:

Children and young adults had higher survival probability.

Middle-aged adults and elderly had lower survival.
